```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('archive/dailyActivity_merged.csv')
df_activity = df
imp_columns = ["Calories", "TotalSteps", "TotalDistance"]
```

```python
df.head()
```

```python
def bar_generator(name, date, format):

    df['date'] = pd.to_datetime(df[date], format=format)

    df.set_index('date', inplace=True)

    plt.figure(figsize=(10, 6))
    plt.bar(df.index, df[name])
    plt.xlabel('Date')
    plt.ylabel(name)
    plt.title(f'Bar Plot of {name}')
    plt.xticks(rotation=45)
    plt.tight_layout()

    plt.show()
```

```python
for name in imp_columns:
    print(name)
    bar_generator(name, 'ActivityDate', '%m/%d/%Y')
```
