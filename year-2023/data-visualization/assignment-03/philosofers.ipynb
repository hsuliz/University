```python
import graphviz

philosophers = [
    {"name": "Thales", "influenced_by": []},
    {"name": "Anaximander", "influenced_by": ["Thales"]},
    {"name": "Anaximenes", "influenced_by": ["Anaximander"]},
    {"name": "Socrates", "influenced_by": ["Parmenides", "Anaxagoras"]},
    {"name": "Plato", "influenced_by": ["Socrates"]},
    {"name": "Aristotle", "influenced_by": ["Plato"]},
    {"name": "René Descartes", "influenced_by": ["Aristotle"]},
    {"name": "Immanuel Kant", "influenced_by": ["René Descartes"]},
    {"name": "Friedrich Nietzsche", "influenced_by": ["Arthur Schopenhauer"]},
    {"name": "Baruch Spinoza", "influenced_by": ["René Descartes"]},
    {"name": "David Hume", "influenced_by": ["John Locke"]},
    {"name": "John Locke", "influenced_by": ["René Descartes"]},
    {"name": "Thomas Hobbes", "influenced_by": ["René Descartes"]},
    {"name": "Jean-Jacques Rousseau", "influenced_by": ["John Locke"]},
    {"name": "Voltaire", "influenced_by": ["John Locke"]},
    {"name": "Bertrand Russell", "influenced_by": ["Immanuel Kant"]},
    {"name": "Simone de Beauvoir", "influenced_by": ["Jean-Paul Sartre"]},
    {"name": "Søren Kierkegaard", "influenced_by": ["Johann Georg Hamann"]},
    {"name": "Michel Foucault", "influenced_by": ["Friedrich Nietzsche"]},
    {"name": "Karl Marx", "influenced_by": ["Georg Wilhelm Friedrich Hegel"]},
    {"name": "Arthur Schopenhauer", "influenced_by": ["Immanuel Kant"]},
    {"name": "Gottfried Leibniz", "influenced_by": ["René Descartes"]},
    {"name": "John Stuart Mill", "influenced_by": ["Jeremy Bentham"]}
]

# Create an empty graph
example_tree = graphviz.Graph(format='svg', graph_attr={'bgcolor': 'white'})

node_color = '#FFD700'
edge_color = '#808080'

for philosopher in philosophers:
    name = philosopher["name"]
    influenced_by = philosopher["influenced_by"]
    if influenced_by:
        for influence in influenced_by:
            example_tree.edge(influence, name, color=edge_color)
    else:
        example_tree.node(name, color=node_color)

example_tree
```
