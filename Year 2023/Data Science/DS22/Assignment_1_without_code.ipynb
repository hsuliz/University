```python
#a)
#Your code:
with open("kc_house_data.csv") as my_file:
    my_file = pd.read_csv(my_file)
    sqr = my_file.get("sqft_living")
    price = my_file.get("price")

    plt.scatter(sqr, price)
    plt.xlabel("living sqft")
    plt.ylabel("price")

    plt.show()


#Hint plt.scatter
```
